# Challenges

1. What was the highest scoring movie in 1996?
2. In what year was the highest rated movie of all time made?
3. What five movies have the most votes ever?
4. What year in the 1960s had the highest average movie rating?

1. Highest scoring movie in 1996

```python
data[data['year'] == 1996].sort_values(by = 'rating', ascending = False).head(1)
```

2. What year was the highest rated movie of all time made

```python
data.groupby(data['year'])['rating'].max().sort_values(ascending = False).head(2)
```

3. Five movies that have the most votes ever

```python
data.sort_values(by = 'votes', ascending = False).head(5)
```

4. Year in the 1960s that had the highest average movie rating

```python
data[(data['year'] >= 1960) & (data['year'] <= 1969)].groupby(data['year']).mean().sort_values(by = 'rating', ascending = False).head(5)
```

# Cleaning data

```python
data[ 'Formatted title'] = data['title'].str.split(' \(').str.get(0)
```

```python
data.head()
```

```python
data['Foormatted length'] = data['length'].str.split().str.get(0).astype('int')
```
